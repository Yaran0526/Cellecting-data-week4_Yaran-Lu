```python
import pandas as pd
```

```python
pd.options.display.max_rows = 200
```

```python
film_df = pd.read_csv('Pudding-Film-Dialogue-Clean.csv', delimiter=",", encoding='utf-8')
```

# Q1:
Sort the proportion_of_dialogue for female characters prior to 2000 in descending order and calculate the average. Then examine the first 20 rows.

```python
film_filter = film_df.loc[(film_df['gender'] == 'woman') & (film_df['release_year'] < 2000)]
```

```python
film_filter.sort_values(by='proportion_of_dialogue', ascending=False).head(20)
```

```python
sorted_dialogue_proportion = film_filter.sort_values(by='proportion_of_dialogue', ascending=False)
average_dialogue_proportion = sorted_dialogue_proportion['proportion_of_dialogue'].mean()
print("Average dialogue proportion:", average_dialogue_proportion)
```

___

# Q2:
Sort the proportion_of_dialogue for female characters after 2000 in descending order and calculate the average. Then examine the first 20 rows.

```python
film_filter2 = film_df.loc[(film_df['gender'] == 'woman') & (film_df['release_year'] > 2000)]
```

```python
film_filter2.sort_values(by='proportion_of_dialogue', ascending=False).head(20)
```

```python
sorted_dialogue_proportion2 = film_filter2.sort_values(by='proportion_of_dialogue', ascending=False)
average_dialogue_proportion2 = sorted_dialogue_proportion2['proportion_of_dialogue'].mean()
print("Average dialogue proportion2:", average_dialogue_proportion2)
```

# Summary:
Before 2000, the average of the proportion of lines for female characters was: 0.08079363587938958, and after 2000, the average of the proportion of lines for female characters was: 0.08211516507322564. From this point of view, there is an upward trend in the proportion of lines for female characters after 2000.

# Q3:
Draw a scatter graph to show the change in the proportion of female characters' lines over the two time periods, with release_year on the X-axis and proportion_of_dialogue on the Y-axis.

```python
ax = sorted_dialogue_proportion.plot(kind='scatter', x='release_year', y='proportion_of_dialogue', color='red')
sorted_dialogue_proportion2.plot(ax=ax, kind='scatter', x='release_year', y='proportion_of_dialogue', color='blue')
```

# Summary:
The scatterplot above roughly depicts the development of the proportion of female character lines in movies over the last 50 years from 1930-2010. It can be seen that before 2000, especially before 1960, the distribution of the data is relatively sparse and the proportion_of_dialogue below 0.2 is more frequent. From 1960 onwards, the sample of data gradually increases, with a more significant cut-off point in 2000, after which the data gradually increase, especially the proportion_of_dialogue above 0.2 is also denser. This graph also represents the fact that after 2000, not only the total number of lines accounted for by female characters is on an upward trend, but also the value of the proportion of individual characters is increasing.
